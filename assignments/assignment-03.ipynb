```python
#1 Write a program that accepts a list from user and print the alternate element of list. 

nums = []

while True:
 input_number = input("Enter some numbers")
 if(input_number == 'q'):
  break
 nums.append(int(input_number))

for i in range(0, len(nums), 2):
    print(nums[i])
```

```python
#2 Write a program that accepts a list from user and reverse it .. don't use the reverse function

numbers = []

while True:
 input_number = input("Enter some numbers")
 if(input_number == 'q'):
  break
 numbers.append(int(input_number))

length = len(numbers) - 1
numbers_reversed = []

while(length >= 0):
 numbers_reversed.append(numbers[length])
 length -= 1

print(f"reversed numbers list {numbers_reversed}")
 
```

```python
#3 Find and display the largest number,  don't use max funciton

numbers = []

while True:
 input_number = input("Enter some numbers")
 if(input_number == 'q'):
  break
 numbers.append(int(input_number))

if(len(numbers) > 0):
 print(numbers)

 maxNumber = numbers[0]
 length = len(numbers) 

 for i in range(0, length):
  testNumber = numbers[i]
  if(testNumber > maxNumber):
   maxNumber = testNumber

 print(maxNumber)
 
```

```python
#4 Write a program that rotates the elements of a list, so that element of first index moves to second index. The element
#at last position moves to first position

numbers = []
reversed_numbers = []


while True:
 input_number = input("Enter some numbers")
 if(input_number == 'q'):
  break
 numbers.append(int(input_number))

if(len(numbers) > 0):
 for num in  reversed(numbers):
  reversed_numbers.append(num)

print(reversed_numbers)
 
```

```python
#6 the code should read in this form mm/dd/yyyy and prints in this form  eg March 10, 2026


input_string = input("Enter some date in form mm/dd/yyyy")

dateArray = input_string.split("/")

monthNumber = dateArray[0]
monthName = ""

if(monthNumber == "01"):
    monthName = "January"
elif(monthNumber == "02"):
    monthName = "February"
elif(monthNumber == "03"):
    monthName = "March"
elif(monthNumber == "04"):
    monthName = "April"
elif(monthNumber == "05"):
    monthName = "May"
elif(monthNumber == "06"):
    monthName = "June"
elif(monthNumber == "07"):
    monthName = "July"
elif(monthNumber == "08"):
    monthName = "August"
elif(monthNumber == "09"):
    monthName = "September"
elif(monthNumber == "10"):
    monthName = "October"
elif(monthNumber == "11"):
    monthName = "November"
else: 
    monthName = "December"

print(f"{monthName} {dateArray[1]}, {dateArray[2]}")
```

```python
#8 find the sum of each row of matrix
matrix = [[2,11,7,12],[5,2,9,15],[8,3,10,42]]

print(matrix)

print(len(matrix))

for i in range(0, len(matrix)):
  row = matrix[i]
  sum = 0
  for r in range(0, len(row)):
   sum += row[r]
  print(f"Sum of row {i+1} = {sum}")
     
```

```python
#9 write program to add two matrices
matrixOne = [[2,11,7,12],[5,2,9,15],[8,3,10,42]]
matrixTwo = [[2,11,7,12],[5,2,9,15],[8,3,10,42]]

rows= len(matrixOne)
cols = len(matrixOne[0])
addedMatrices = []


for i in range(rows):
    row = []
    for j in range(cols):
        row.append(matrixOne[i][j] + matrixTwo[i][j])
    addedMatrices.append(row)

print(addedMatrices)



```
