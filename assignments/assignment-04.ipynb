```python
#4 create frequency dictionary

numbers = [1,2,2,3,3,4,4,3,4,4,4,44,4,15,6]

frequency_dictionary = {}
for num in numbers:
    current = num
    if(num in frequency_dictionary.keys()):
        frequency_dictionary[num] += 1
    else:
        frequency_dictionary[num] = 1
    
for key, value in frequency_dictionary.items():
    print(f"{key} is present {value} times")


```

```python
#5 word counter, take a sentence input and count the frequency of each word.
#weather is good but has clouds clouds but sunny but is good

input_string = input("Enter a String to count the frequency of each word")
input_string_array = input_string.split(" ")

frequency_dictionary = {}
for word in input_string_array:
    current = word
    if(word in frequency_dictionary.keys()):
        frequency_dictionary[word] += 1
    else:
        frequency_dictionary[word] = 1
    
for key, value in frequency_dictionary.items():
    print(f"{key} is present {value} times")
```

```python
#6 Student marks and average

marks = {
    "Maths": 80,
    "English" : 70,
    "Physics": 90
}

total_num_of_subjects =0
total_marks = 0
for total_marks in marks.values():
    total_num_of_subjects +=1
    total_marks += total_marks

avg = total_marks /total_num_of_subjects

print(f"Avg marks is  {avg}")


```

```python
#7 find the highest scoring student

student = {
    "Ali": 85,
    "Sara": 92,
    "Ahmed":78,
    "Zahra": 95
}

highest =0
person = ""

for name, marks in student.items():
    if(marks >= highest):
        highest = marks
        person = name

print(f"{person} got the highest {highest}")
```

```python
#8 merge dictionaries. 

dictionary_one = {
    "a": 10, 
    "b": 20
}
dictionary_two = {
    "c": 10, 
    "d": 20
}


# merge dictionaries
for key, value in dictionary_two.items():
    dictionary_one[key] = value


print(dictionary_one)
```
