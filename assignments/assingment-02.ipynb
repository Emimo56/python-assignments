```python
quantity_purchased = int(input("Enter quantity purchased"))
price_per_item = 100
discount_percentage = 10 /100

cost = price_per_item * quantity_purchased
cost_after_discount = 0

if(cost > 1000):
    discount = cost * discount_percentage
    cost_after_discount = cost - discount
    print("Cost after discount", cost_after_discount)
else:
    cost_after_discount = cost
    print("Cost after discount", cost_after_discount)
```

```python
classes_attended = int(input("Enter your classes attended"))
classes_held = int(input("Enter your classes held"))

attendance_needed = 75.0

actual_attendance = (classes_attended / classes_held) * 100

print("Percentage of class attended ", actual_attendance);

if(actual_attendance < attendance_needed):
    medical_condition =  input("Do you have medical condition");
    if(medical_condition):
        print("Allowed to sit in Exam because of medical condition")
    else:
        print("Not allowed to sit in exam")
else:
    print("Allowed to sit in exam")

```
